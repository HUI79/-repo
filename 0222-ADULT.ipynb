```python
import pandas as pd
import numpy as np

# 讀取CSV文件
df = pd.read_csv('adult.csv')

# 選取"Income"欄位
income_data = df['Income']

# 使用NumPy進行統計計算
mean_income = np.mean(income_data)
std_income = np.std(income_data)
median_income = np.median(income_data)

# 顯示結果
print("平均值：", mean_income)
print("標準差：", std_income)
print("中位數：", median_income)
```
